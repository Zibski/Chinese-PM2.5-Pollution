```python
import os

directory = os.getcwd() + '/input/'
inputs = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".csv"):
        inputs.append(os.path.join(directory, filename))
    else:
        continue
```

```python
import pandas as pd
df_Shanghai = pd.read_csv(inputs[0])
df_Beijing = pd.read_csv(inputs[1])
df_Shenyang = pd.read_csv(inputs[2])
df_Chengdu = pd.read_csv(inputs[3])
df_Guangzhou = pd.read_csv(inputs[4])
```

```python
df_Shanghai['PM_mean'] = df_Shanghai[['PM_Jingan', 'PM_US Post', 'PM_Xuhui']].mean(axis=1)
df_Beijing['PM_mean'] = df_Shanghai[['PM_Jingan', 'PM_US Post', 'PM_Xuhui']].mean(axis=1)
df_Shenyang['PM_mean'] = df_Shanghai[['PM_Jingan', 'PM_US Post', 'PM_Xuhui']].mean(axis=1)
df_Chengdu['PM_mean'] = df_Shanghai[['PM_Jingan', 'PM_US Post', 'PM_Xuhui']].mean(axis=1)
df_Guangzhou['PM_mean'] = df_Shanghai[['PM_Jingan', 'PM_US Post', 'PM_Xuhui']].mean(axis=1)
```
